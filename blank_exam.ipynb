## Question 1

For a 2-qubit system in Qiskit, which Pauli label applies Z on qubit 1 and X on qubit 0?

**A.** `'XZ'`  
**B.** `'ZX'`  
**C.** `'ZI'`  
**D.** `'IX'`

<details>
<summary>Click to reveal Answer & Explanation</summary>

**Correct Answer: B**

**Explanation:**  
Qiskit Pauli strings are right-to-left by qubit index (rightmost char is qubit 0).  
So `Z` on qubit 1 and `X` on qubit 0 is `'ZX'`.  
- `'XZ'` would mean X on qubit 1 and Z on qubit 0.

</details>

## Question 2

Which statement is correct for the commutator `[X, Z] = XZ - ZX`?

**A.** `[X, Z] = 0`  
**B.** `[X, Z] = 2iY`  
**C.** `[X, Z] = -2iY`  
**D.** `[X, Z] = I`

<details>
<summary>Click to reveal Answer & Explanation</summary>

**Correct Answer: C**

**Explanation:**  
Using Pauli algebra: `XZ = -iY` and `ZX = iY`, so `XZ - ZX = -2iY`.  
- **A** is false (they anti-commute).  
- **B** has the wrong sign.  
- **D** is unrelated.

</details>

## Question 8

Which `SparsePauliOp` expression is dimensionally valid for a 3-qubit Hamiltonian?

**A.** `SparsePauliOp.from_list([('ZZ', 0.7), ('XII', 0.2)])`  
**B.** `SparsePauliOp.from_list([('ZIZ', 0.7), ('XII', 0.2)])`  
**C.** `SparsePauliOp.from_list([('Z', 0.7), ('XII', 0.2)])`  
**D.** `SparsePauliOp.from_list([('IIII', 1.0), ('XII', 0.2)])`

<details>
<summary>Click to reveal Answer & Explanation</summary>

**Correct Answer: B**

**Explanation:**  
All Pauli strings in one `SparsePauliOp` must have consistent length equal to qubit count.  
- **B** uses only 3-character strings.  
- **A** and **C** mix 2/1 with 3.  
- **D** mixes 4 with 3.

</details>
